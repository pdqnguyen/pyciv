```python
board = main((160, 90))
SQRT3 = sqrt(3)
width, height = (1280, 720)
max_x = ceil(width / (board.shape[0] + 0.5) / SQRT3)
max_y = ceil(height / (1.5 * board.shape[1] + 0.5))
radius = min(max_x, max_y)
# width = board.shape[0] * radius * 1.5 + radius / 2.0
# height = (board.shape[1] + .5) * radius * SQRT3 + 1

def get_base_color(tile):
    base_color, _ = board._get_colors(tile)
    rgb = to_rgb(base_color)
    r, g, b = [int(x * 255) for x in rgb]
    return pygame.Color(*[int(x * 255) for x in rgb])

def get_feature_color(tile):
    _, feature_color = board._get_colors(tile)
    if feature_color:
        rgb = to_rgb(feature_color)
        r, g, b = [int(x * 255) for x in rgb]
        return pygame.Color(r, g, b)
    else:
        return None

surface = pygame.Surface((width, height))
def draw(surface, board, radius):
    for tile in board:
        base_color = get_base_color(tile)
        feature_color = get_feature_color(tile)
        if tile.y % 2 == 0:
            left = SQRT3 * radius * (tile.x)
        else:
            left = SQRT3 * radius * (tile.x + 0.5)
        top = 1.5 * radius * (tile.y)
        points = [
            (0.5 * SQRT3 * radius, 0),
            (SQRT3 * radius, 0.5 * radius),
            (SQRT3 * radius, 1.5 * radius),
            (0.5 * SQRT3 * radius, 2 * radius),
            (0, 1.5 * radius),
            (0, 0.5 * radius)
        ]
        circle_pos = (int(left + 0.5 * SQRT3 * radius), int(height - (top + radius)))
        circle_r = int(0.5 * radius)
        points = [(x + left, height - (y + top)) for (x, y) in points]
        pygame.draw.polygon(surface, base_color, points)
        pygame.draw.polygon(surface, pygame.Color(0, 0, 0), points, 2)
        if feature_color:
            pygame.draw.circle(surface, feature_color, circle_pos, circle_r)
        

try:
    pygame.init()
    fpsClock = pygame.time.Clock()

    window = pygame.display.set_mode((width, height), 1)
    from pygame.locals import QUIT, MOUSEBUTTONDOWN

    #Leave it running until exit
    while True:
        for event in pygame.event.get():
            if event.type == QUIT:
                pygame.quit()
                sys.exit()
            if event.type == MOUSEBUTTONDOWN:
                print(event.pos)
        draw(surface, board, radius)
        window.blit(surface, (0, 0))
        pygame.display.update()
        fpsClock.tick( 10 )
finally:
    pygame.quit()
```
